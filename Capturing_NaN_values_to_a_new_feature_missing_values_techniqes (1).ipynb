#Capturing NaN values with a new feature:It works well if the data are not missing completely at random

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df['Age'].isnull().sum()
```

```python
df['Age_nan']=np.where(df['Age'].isnull(),1,0)
```

```python
df.head()
```

```python
df.Age.median()
```

```python
df['Age'].fillna(df.Age.median,inplace=True)
```

```python
df['Age_nan'].unique()
```
