#Arbitrary value imputation:It consist of replacing NaN values with an arbitrary value

```python
def impute_nan(df,variable):
  df[variable+"_zero"]=df[variable].fillna(0)
  df[variable+"_hundred"]=df[variable].fillna(100)
```

```python
impute_nan(df,'Age')
```
