```python
df['X0'].unique()
```

```python
X=df[['X0','X1','X2']]
X.head()
```

```python
for feature in X.columns[:]:
  print(feature, ":" , len(X[feature].unique()))
```

```python
X['X0'].value_counts()
```

```python
Dict=X['X0'].value_counts().to_dict()
Dict
```

```python
X['X0']=X['X0'].map(Dict)
X.head()
```

```python
#One-Hot Encoding: This technique creates a binary column for each category in the categorical variable. Each observation receives a 1 in the column corresponding to its category, and 0s in all other columns. One-hot encoding is suitable for nominal categorical variables (categories with no inherent order).

'''Label Encoding: Label encoding assigns a unique integer to each category in the variable. It's suitable for ordinal categorical variables (categories with a specific order), but it can introduce unintended ordinality where none exists.

Ordinal Encoding: This is similar to label encoding but more explicit in handling ordinal variables. You manually assign ranks or orders to the categories. It's more appropriate when there's a clear order among the categories.

Binary Encoding: Binary encoding converts each category into binary digits. First, the categories are converted to integers using label encoding, then these integers are converted into binary code, and the digits are split into separate columns. It reduces the dimensionality compared to one-hot encoding.

Frequency Encoding: In frequency encoding, each category is replaced with the frequency of that category in the dataset. It can be useful for high-cardinality categorical variables.

Target Encoding (Mean Encoding): In target encoding, each category is replaced with the mean of the target variable for that category. It's useful for classification problems and can sometimes improve performance, but it's prone to overfitting, especially on smaller datasets.

Hashing Encoding: This technique hashes the categorical values into a specified number of bins, which can be particularly useful when dealing with high cardinality categorical variables.

Embedding: Commonly used in neural network-based models, embedding maps categorical variables to a lower-dimensional continuous space. It's effective for dealing with high cardinality categorical variables and capturing complex relationships between categories.

Each of these techniques has its advantages and disadvantages, and the choice depends on factors such as the nature of the data, the algorithm being used, and the specific goals of the analysis.'''
```

Target Guided/Mean Ordinal Encoding


```python
#1.Ordering the labels according to the target
#2.Replace the labels by being joint probability of 0 or 1
```

```python
df=pd.read_csv("titanic.csv",usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna("Missing",inplace=True)
df.head()
```

```python
df['Cabin'].astype(str).str[0]
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df
```

```python
df.groupby('Cabin')['Survived'].mean().sort_values()
```

```python
ordinal_labels=df.groupby('Cabin')['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels1={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels1
```

```python
df['Cardinal_labels']=df['Cabin'].map(ordinal_labels1)
df.head()
```

#Mean Encoding


```python
mean_encoding=df.groupby('Cabin')['Survived'].mean().to_dict()
mean_encoding
```

```python
df['Cabin_mean_encoding']=df['Cabin'].map(mean_encoding)
df.head()
```
